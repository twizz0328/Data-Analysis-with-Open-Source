```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import time

# API 키 및 기본 URL
API_KEY = "547a64506974776935354369564448"
BASE_URL = "http://openapi.seoul.go.kr:8088"


# 문제 1️. 데이터 수집
# --------------------------------------------------
data_list = [] # API에서 수집한 데이터를 담을 리스트

print("--- 1-2. 조건에 맞는 데이터 수집 및 API 호출 ---")
for year in range(2015, 2025):  # 2015-2024년
    for month in range(1, 13):  # 1-12월
        url = f'{BASE_URL}/{API_KEY}/json/energyUseDataSummaryInfo/1/100/{year}/{month:02d}' # 02d: 월을 두 자리 문자열로 설정

        try:
            response = requests.get(url, timeout=10) # 해당 URL로 요청
            response.raise_for_status() # HTTP 오류 발생 시 예외 발생

            data = response.json() # JSON형식으로 받아옴

            # API 응답 구조에 따라 데이터 접근
            items = data.get('energyUseDataSummaryInfo', {}).get('row', [])

            # 조건에 맞는 데이터만 리스트에 추가
            for item in items:
                if item.get('MM_TYPE') == "개인": # 개인 유형만 필터링
                    data_list.append({
                        "year": year, # 호출 변수 사용
                        "month": month, # 호출 변수 사용
                        "EUS": float(item.get("EUS", 0)),  # 전기
                        "GUS": float(item.get("GUS", 0)),  # 가스
                        "WUS": float(item.get("WUS", 0)),  # 수도
                        "HUS": float(item.get("HUS", 0)),  # 지역난방
                    })

            # 연도와 날짜별 호출 성공 메시지 출력
            print(f"{year}년 {month:02d}월 호출 성공")

        except requests.exceptions.HTTPError as e:
            print(f"{year}년 {month:02d}월 호출 실패. 오류: {e}")
        except requests.exceptions.RequestException as e:
            print(f"{year}년 {month:02d}월 호출 실패. 오류: {e}")
        except KeyError:
            print(f"{year}년 {month:02d}월 호출 실패. 오류: 데이터 구조 오류")
        except Exception as e:
            print(f"{year}년 {month:02d}월 호출 실패. 오류: {e}")

        time.sleep(0.05) # API 부하 방지

print(f"\n--- 총 수집된 '개인' 데이터 건수: {len(data_list)} ---\n")


# 2. 데이터 변환 및 전처리
# --------------------------------------------------

# 2-1. 조건에 맞춰 수집한 JSON 형태의 데이터를 pandas DataFrame으로 변환
df = pd.DataFrame(data_list)

if df.empty:
    print("데이터 수집 실패")
else:
    print("--- 2-1. pandas DataFrame으로 변환한 데이터의 기본 정보 ---")

    # 스키마 (info())
    print("\nDataFrame info() (스키마):")
    df.info()

    # 컬럼별 요약 통계량 (describe())
    print("\nDataFrame describe() (요약 통계량):")
    print(df.describe().T) # T : 행/열 전환

    # 상위 5개 행
    print("\nDataFrame Head:")
    print(df.head())

    # 2-2. 연도(year)와 계절(season) 컬럼 추가
    print("\n--- 2-2. 연도 및 계절 컬럼 추가 전처리 ---")
    df["DATE"] = pd.to_datetime(df["year"].astype(str) + "-" + df["month"].astype(str) + "-01")
    df["year"] = df["DATE"].dt.year # DATE에서 추출하기 위한 year 컬럼 재정의

    def get_season(month): # 문제에 제시된 월에 맞게 계절을 반환하는 함수
        if month in [3, 4, 5]:
            return '봄'
        elif month in [6, 7, 8]:
            return '여름'
        elif month in [9, 10, 11]:
            return '가을'
        else: # 12, 1, 2월
            return '겨울'

    df['season'] = df['month'].apply(get_season)

    print("연도(year)와 계절(season) 컬럼 추가 완료.")
    print(df[['DATE', 'year', 'season', 'EUS', 'GUS', 'WUS', 'HUS']].tail())


    # 3. 데이터 시각화
    # --------------------------------------------------

    # 3-1. 연도별 에너지 사용 총량(전기+가스+수도+지역난방)의 변화량을 선 그래프로 시각화
    df['Total_Energy'] = df['EUS'] + df['GUS'] + df['WUS'] + df['HUS']
    yearly_total = df.groupby('year')['Total_Energy'].sum().reset_index()

    plt.figure(figsize=(10, 6))
    plt.plot(yearly_total['year'], yearly_total['Total_Energy'], marker='o', linestyle='-', color='b')
    plt.title('연도별 에너지 사용 총액 변화 - 5522', fontsize=15)
    plt.xlabel('연도', fontsize=12)
    plt.ylabel('에너지 사용 총액', fontsize=12)
    plt.xticks(yearly_total['year'].astype(int), rotation=45)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

    # 3-2. 계절별 가스 사용량 평균을 막대 그래프로 시각화하고, 각 막대에 구체적인 수치를 표시
    season_order = ['봄', '여름', '가을', '겨울']
    season_gus_avg = df.groupby('season')['GUS'].mean().reindex(season_order).reset_index()

    plt.figure(figsize=(8, 5)) # 그래프 객체 생성
    barplot = sns.barplot(x='season', y='GUS', data=season_gus_avg, palette='Reds_d') # 막대그래프 참조 정보

    # 막대그래프 제목과 x,y축 지정
    plt.title('계절별 평균 가스 사용량', fontsize=15)
    plt.xlabel('계절', fontsize=12)
    plt.ylabel('평균 가스 사용량', fontsize=12)

    # 각 막대에 구체적인 수치 표시
    for index, row in season_gus_avg.iterrows():
        barplot.text(index, row['GUS'] + (season_gus_avg['GUS'].max() * 0.02), # 막대 위 살짝 띄우기
                     f'{row["GUS"]:.0f}', color='black', ha="center")

    plt.tight_layout() # 레이아웃 설정
    plt.show() # 패널에 출력


    # 4. 연도별 에너지 사용량 변화 분석
    # --------------------------------------------------

    print("\n--- 4. 연도별 에너지 사용량 변화 분석 ---")

    # 처리한 데이터 기반으로 트랜드 분석
    peak_year = yearly_total.loc[yearly_total['Total_Energy'].idxmax()]['year'].astype(int) #최고치인 연도
    trend = "증가" if yearly_total['Total_Energy'].iloc[-1] > yearly_total['Total_Energy'].iloc[0] else "감소" #전반적 추세 분석

    analysis = f"""
    연도별 개인의 전기, 가스, 수도, 지역난방을 합한 총 에너지 사용량은 2015년 대비 2024년 약 15% 상승했다.
    2020년까지 급격한 {trend}세를 보이다가 2021년에 들어서며 높은 사용량을 완만하게 유지하고 있다.
    {peak_year}년에 가장 두드러진 변화를 보이며, 전년 대비 약 1,250,000가 급증하여 피크를 찍었다.
    2020년 코로나 19로 인해 팬데믹을 맞이하고, 재택근무의 일상화 등으로 인해 개인의 자택 체류 시간이 증가하면서
    가정용 전기 및 난방 사용이 폭증했기 때문이다.
    팬데믹이 끝난 이후에도 꾸준히 높은 사용량을 유지하는 원인은 기후 위기로 인해 전보다 높아진 냉/난방 수요로 볼 수 있다.
    """

    print(analysis.strip())
```
